```python
#split the dataset 
x=df.drop(["Churn"],axis=1)
y=df["Churn"] 
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,shuffle=True,test_size=0.2,random_state=42)
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
model1=DecisionTreeClassifier(criterion= "log_loss", max_depth= 40, max_features="sqrt", min_samples_leaf= 5, min_samples_split= 40, splitter= "random")
```

```python
parmeter={


"criterion":['gini', 'entropy', 'log_loss'] ,
"splitter": ['best', 'random'] ,
"max_depth" : [20,25,30,40],
"min_samples_split":[20,10,30,40],
"min_samples_leaf" :[1,2,3,4,5,6,7,8,9],

"max_features":['auto', 'sqrt', 'log2'] 
 }
```

```python
gred_search=GridSearchCV(model1,parmeter,cv=5,n_jobs=-1)
gred_search.fit(x_train,y_train)
```

```python
print(gred_search.best_params_)
```

```python
model2=RandomForestClassifier(bootstrap= True, max_depth= 20, max_features= 'sqrt', min_samples_leaf= 2, min_samples_split= 40, n_estimators= 20)
```

```python
Parmeter2={
"n_estimators":[5,10,20,30],
"max_depth" : [20,25,30,40],
"min_samples_split":[20,10,30,40],
"min_samples_leaf" :[1,2,3,4,5,6,7,8,9],
"bootstrap":[True,False],
"max_features":['auto', 'sqrt'], 
}
```

```python
gred_search2=GridSearchCV(model2,Parmeter2,cv=5,n_jobs=-1)
gred_search2.fit(x_train,y_train)
```

```python
print(gred_search2.best_params_)
```

```python
model3=XGBClassifier()
```

```python
voot=VotingClassifier(
    estimators=[ 
        ("mo1",model1),
        ("mo2",model2),
        ("mo3",model3)
    ],
    voting="hard"
)
```

```python
voot.fit(x_train,y_train)
```

```python
k=KFold(n_splits=5,shuffle=True,random_state=42)
cross_val_score(voot,x,y,cv=k)
```

```python
y_pred=voot.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
cm=confusion_matrix(y_test,y_pred)
cmd=ConfusionMatrixDisplay(cm)                     
cmd.plot()
```
